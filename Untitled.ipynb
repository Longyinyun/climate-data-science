```python
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset

file1 = r"C:\Users\19146\Desktop\data\pr_Amon_CESM1-CAM5_rcp85_r1i1p1_200601-210012.nc"     #降水
file2 = r"C:\Users\19146\Desktop\data\tas_Amon_CESM1-CAM5_rcp85_r1i1p1_200601-210012.nc"    #气温

ds1 = Dataset(file1,'r')
ds2 = Dataset(file2,'r')

pr = ds1.variables['pr'][:]
tas = ds2.variables['tas'][:]

# site New York
# lon -74, lat 40.7
lon = ds1.variables['lon'][:]
lat = ds1.variables['lat'][:]

lon_NY_index = int((360-74)/2.5)                       #经度
lat_NY_index = int(40.7 / 2) + 45                      #纬度

tas_NY = tas[:,lat_NY_index,lon_NY_index]             
pr_NY = pr[:,lat_NY_index,lon_NY_index]                     #时间序列

from scipy.stats import pearsonr
p = pearsonr(tas_NY,pr_NY)

lag = 10
p1 = pearsonr(tas_NY[lag:],pr_NY[:-lag])

# start here
Data = np.hstack((tas_NY.reshape(-1,1),pr_NY.reshape(-1,1)))    #reshape把行向量转成列向量


varNames = ['tas','pr']              #加变量
nSignals=2
nSignals = Data.shape[1]
nData = Data.shape[0]

nBins = 10         #别改

# classifiy data
binEdges = np.full([nSignals,nBins],np.nan)
minEdge = np.full([nSignals],np.nan)
maxEdge = np.full([nSignals],np.nan)
classifiedData = np.full([nData,nSignals],np.nan)

for s in range(nSignals):
    minEdge[s] = np.min(Data[:,s])
    maxEdge[s] = np.max(Data[:,s])
    binEdges_tmp = np.linspace(minEdge[s],maxEdge[s],num=nBins+1)
    binEdges[s,0:nBins] = binEdges_tmp[1:]

for s in range(nSignals):
    smat = Data[:,s].copy()
    cmat = np.full([nData],np.nan)
    for e in range(nBins):
        if np.where(smat<=binEdges[s,e])[0].size > 0:
            cmat[np.where(smat<=binEdges[s,e])[0]] = e
            smat[np.where(smat<=binEdges[s,e])[0]] = maxEdge[s] + 999
    classifiedData[:,s] = cmat

# calculate Transfer entropy
nTuples = nData - 1 # ?
# matrix
HYwYf = np.full([nSignals,nSignals],np.nan)
HXtYw = np.full([nSignals,nSignals],np.nan)
HYw = np.full([nSignals,nSignals],np.nan)
HXtYwYf = np.full([nSignals,nSignals],np.nan)
T = np.full([nSignals,nSignals],np.nan)
I = np.full([nSignals,nSignals],np.nan)
Tplus = np.full([nSignals],0.0)
Tminus = np.full([nSignals],0.0)
Tnet = np.full([nSignals],0.0)

for sX in range(nSignals):                     #sX是原因，sY是结果  循环出四种可能：降水—>降水，降水—>温度，温度—>温度，温度—>降水
    for sY in range(nSignals):
        #
        tupleMat = np.full([nTuples,3],np.nan)             #初始化
        # Xt
        tupleMat[:,0] = classifiedData[0:nTuples,sX]       #X的历史
        # Yw
        tupleMat[:,1] = classifiedData[0:nTuples,sY]       #Y的历史
        # Yf
        tupleMat[:,2] = classifiedData[1:nTuples+1,sY]     #Y的当前
        
        # Xt,Yw,Yf, pXtYwYf
        C = np.full([nBins,nBins,nBins],0)
        for i in range(len(tupleMat)):
            C[int(tupleMat[i,0]),int(tupleMat[i,1]),int(tupleMat[i,2])] = \
            C[int(tupleMat[i,0]),int(tupleMat[i,1]),int(tupleMat[i,2])] + 1
        pXtYwYf = (C + 1e-20)/np.sum(C)               #三维联合概率密度     + 1e-20是为了避免0值
        
        #
        pYw = np.sum(np.sum(pXtYwYf,axis=0),axis=1)         #Y历史的概率
        pXtYw = np.sum(pXtYwYf,axis=2)                      #X历史和Y历史的联合概率
        pYwYf = np.sum(pXtYwYf,axis=0)                      #Y历史和Y当前的联合概率
        
        # p*log2p     算信息熵
        HXtYwYf[sX,sY] = -1.0*np.sum(np.sum(np.sum(pXtYwYf*np.log2(pXtYwYf))))
        HXtYw[sX,sY] = -1.0*np.sum(np.sum(pXtYw*np.log2(pXtYw)))
        HYwYf[sX,sY] = -1.0*np.sum(np.sum(pYwYf*np.log2(pYwYf)))
        HYw[sX,sY] = -1.0*np.sum(pYw*np.log2(pYw))
        
        # transfer                 传递熵
        T[sX,sY] = HXtYw[sX,sY] + HYwYf[sX,sY] - HYw[sX,sY] - HXtYwYf[sX,sY]
        # gross information production
        Tplus[sX] = T[sX,sY] + Tplus[sX]         #产生信息的总量
        # gross information consumption
        Tminus[sY] = T[sX,sY] + Tminus[sY]       #吸收信息的总量
        # net information gain                   #信息增益
        Tnet[sX] = Tnet[sX]  + T[sX,sY]
        Tnet[sY] = Tnet[sY]  - T[sX,sY]
print(T)



import pandas as pd

Gephi_node = []
for sX in range(nSignals):
    Gephi_node.append([varNames[sX],varNames[sX],Tplus[sX],Tminus[sX],Tnet[sX]])
Gephi_node_df = pd.DataFrame(data=Gephi_node,columns=['Id','Label','Tplus','Tminus','Tnet'])
Gephi_node_df.to_csv('Gephi_node.csv',encoding='utf-8',index=False)

Gephi_edge = []
for sX in range(nSignals):
    for sY in range(nSignals):
        if sX != sY:
            Gephi_edge.append([varNames[sX],varNames[sY],T[sX,sY]])
Gephi_edge_df = pd.DataFrame(data=Gephi_edge, columns = ['Source', 'Target', 'T'])
Gephi_edge_df.to_csv('Gephi_edge.csv',encoding='utf-8',index=False)

# whole network (more than two variables) analysis
```

```python
# whole network (more than two variables) analysis
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset

file1 = r"C:\Users\19146\Desktop\data\pr_Amon_CESM1-CAM5_rcp85_r1i1p1_200601-210012.nc"     #降水
file2 = r"C:\Users\19146\Desktop\data\tas_Amon_CESM1-CAM5_rcp85_r1i1p1_200601-210012.nc"    #气温
file3 = r"C:\Users\19146\Desktop\data\sfcWind_Amon_CESM1-CAM5_rcp85_r1i1p1_200601-210012.nc"    #风速
file4 = r"C:\Users\19146\Desktop\data\ps_Amon_CESM1-CAM5_rcp85_r1i1p1_200601-210012.nc"      #气压
ds1 = Dataset(file1,'r')
ds2 = Dataset(file2,'r')
ds3 = Dataset(file3,'r')
ds4 = Dataset(file4,'r')

pr = ds1.variables['pr'][:]
tas = ds2.variables['tas'][:]
wind = ds3.variables['sfcWind'][:]
ps = ds4.variables['ps'][:]

# site Shanghai
# 121.46 30.92
lat_index = int((30.92/2) + 45)
lon_index = int((121.46)/2.5)

#时间序列
tas_SH = tas[:,lat_NY_index,lon_NY_index]             
pr_SH = pr[:,lat_NY_index,lon_NY_index]                     
wind_SH = wind[:,lat_NY_index,lon_NY_index] 
ps_SH = ps[:,lat_NY_index,lon_NY_index]    

'''
from scipy.stats import pearsonr
p = pearsonr(tas_SH,pr_SH,wind_SH )

lag = 10
p1 = pearsonr(tas_SH[lag:],pr_SH[:-lag])
'''
# start here
Data = np.hstack((tas_SH.reshape(-1,1),pr_SH.reshape(-1,1),wind_SH.reshape(-1,1),ps_SH.reshape(-1,1)))    #reshape把行向量转成列向量


varNames = ['tas','pr','wind','ps']              #加变量
nSignals=4
nSignals = Data.shape[1]
nData = Data.shape[0]

nBins = 10         #别改

# classifiy data
binEdges = np.full([nSignals,nBins],np.nan)
minEdge = np.full([nSignals],np.nan)
maxEdge = np.full([nSignals],np.nan)
classifiedData = np.full([nData,nSignals],np.nan)

for s in range(nSignals):
    minEdge[s] = np.min(Data[:,s])
    maxEdge[s] = np.max(Data[:,s])
    binEdges_tmp = np.linspace(minEdge[s],maxEdge[s],num=nBins+1)
    binEdges[s,0:nBins] = binEdges_tmp[1:]

for s in range(nSignals):
    smat = Data[:,s].copy()
    cmat = np.full([nData],np.nan)
    for e in range(nBins):
        if np.where(smat<=binEdges[s,e])[0].size > 0:
            cmat[np.where(smat<=binEdges[s,e])[0]] = e
            smat[np.where(smat<=binEdges[s,e])[0]] = maxEdge[s] + 999
    classifiedData[:,s] = cmat

# calculate Transfer entropy
nTuples = nData - 1 # ?
# matrix
HYwYf = np.full([nSignals,nSignals],np.nan)
HXtYw = np.full([nSignals,nSignals],np.nan)
HYw = np.full([nSignals,nSignals],np.nan)
HXtYwYf = np.full([nSignals,nSignals],np.nan)
T = np.full([nSignals,nSignals],np.nan)
I = np.full([nSignals,nSignals],np.nan)
Tplus = np.full([nSignals],0.0)
Tminus = np.full([nSignals],0.0)
Tnet = np.full([nSignals],0.0)

for sX in range(nSignals):                     #sX是原因，sY是结果  
    for sY in range(nSignals):
        #
        tupleMat = np.full([nTuples,3],np.nan)             #初始化
        # Xt
        tupleMat[:,0] = classifiedData[0:nTuples,sX]       #X的历史
        # Yw
        tupleMat[:,1] = classifiedData[0:nTuples,sY]       #Y的历史
        # Yf
        tupleMat[:,2] = classifiedData[1:nTuples+1,sY]     #Y的当前
        
        # Xt,Yw,Yf, pXtYwYf
        C = np.full([nBins,nBins,nBins],0)
        for i in range(len(tupleMat)):
            C[int(tupleMat[i,0]),int(tupleMat[i,1]),int(tupleMat[i,2])] = \
            C[int(tupleMat[i,0]),int(tupleMat[i,1]),int(tupleMat[i,2])] + 1
        pXtYwYf = (C + 1e-20)/np.sum(C)               #三维联合概率密度     + 1e-20是为了避免0值
        
        #
        pYw = np.sum(np.sum(pXtYwYf,axis=0),axis=1)         #Y历史的概率
        pXtYw = np.sum(pXtYwYf,axis=2)                      #X历史和Y历史的联合概率
        pYwYf = np.sum(pXtYwYf,axis=0)                      #Y历史和Y当前的联合概率
        
        # p*log2p     算信息熵
        HXtYwYf[sX,sY] = -1.0*np.sum(np.sum(np.sum(pXtYwYf*np.log2(pXtYwYf))))
        HXtYw[sX,sY] = -1.0*np.sum(np.sum(pXtYw*np.log2(pXtYw)))
        HYwYf[sX,sY] = -1.0*np.sum(np.sum(pYwYf*np.log2(pYwYf)))
        HYw[sX,sY] = -1.0*np.sum(pYw*np.log2(pYw))
        
        # transfer                 传递熵
        T[sX,sY] = HXtYw[sX,sY] + HYwYf[sX,sY] - HYw[sX,sY] - HXtYwYf[sX,sY]
        # gross information production
        Tplus[sX] = T[sX,sY] + Tplus[sX]         #产生信息的总量
        # gross information consumption
        Tminus[sY] = T[sX,sY] + Tminus[sY]       #吸收信息的总量
        # net information gain                   #信息增益
        Tnet[sX] = Tnet[sX]  + T[sX,sY]
        Tnet[sY] = Tnet[sY]  - T[sX,sY]
print(T)



import pandas as pd

Gephi_node = []
for sX in range(nSignals):
    Gephi_node.append([varNames[sX],varNames[sX],Tplus[sX],Tminus[sX],Tnet[sX]])
Gephi_node_df = pd.DataFrame(data=Gephi_node,columns=['Id','Label','Tplus','Tminus','Tnet'])
Gephi_node_df.to_csv('Gephi_node3.csv',encoding='utf-8',index=False)

Gephi_edge = []
for sX in range(nSignals):
    for sY in range(nSignals):
        if sX != sY:
            Gephi_edge.append([varNames[sX],varNames[sY],T[sX,sY]])
Gephi_edge_df = pd.DataFrame(data=Gephi_edge, columns = ['Source', 'Target', 'T'])
Gephi_edge_df.to_csv('Gephi_edge3.csv',encoding='utf-8',index=False)

```
